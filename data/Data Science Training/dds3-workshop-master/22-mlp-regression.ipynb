# MLP Regression

MLP's are supervised algorithms. We need a target so we can generate the output error.

That means we can also use it for regression. The error is the error between the target function and the one predicted by the model.

```python
from sklearn.neural_network import MLPRegressor
import numpy as np
import matplotlib.pyplot as plt
```

Let's create a simple function to learn and play about with an `MLPRegressor`.

```python
x = np.arange(0.0, 1, 0.01).reshape(-1, 1)
y = np.sin(2 * np.pi * x).ravel()

clf = MLPRegressor(
    hidden_layer_sizes=(10,),  activation='relu', solver='adam', alpha=0.001, batch_size='auto',
    learning_rate='constant', learning_rate_init=0.01, power_t=0.5, max_iter=1000, shuffle=True,
    random_state=9, tol=0.0001, verbose=False, warm_start=False, momentum=0.9, nesterovs_momentum=True,
    early_stopping=False, validation_fraction=0.1, beta_1=0.9, beta_2=0.999, epsilon=1e-08)

clf.fit(x, y)

fig, axarr = plt.subplots(1, 2, figsize=(11, 5))
ax = axarr[0]
ax.plot(range(1, len(clf.loss_curve_) + 1), clf.loss_curve_)
ax.set_title("Errors for each iteration")
ax.set_xlabel("Iteration")
ax.set_ylabel("MLP error")

ax = axarr[1]
test_x = np.arange(0.0, 1, 0.05).reshape(-1, 1)
test_y = clf.predict(test_x)
ax.scatter(x, y, s=1, c='b', marker="s", label='real')
ax.scatter(test_x,test_y, s=10, c='r', marker="o", label='NN Prediction')
ax.set_title("MLP Fitted Model")
plt.show()
```

_Note: We have no concept of time here. All points are being fed into the MLP at the same time._
